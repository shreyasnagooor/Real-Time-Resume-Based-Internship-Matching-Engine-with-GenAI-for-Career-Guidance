```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()


def preprocess_text(text):
    if not isinstance(text, str):
        return ''

    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()


    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python

df1['Cleaned_Job_Description'] = df1['Cleaned_Job_Description'].apply(preprocess_text)
df2['Cleaned_Text'] = df2['Cleaned_Text'].apply(preprocess_text)

print("Job Descriptions After Preprocessing:")
print(df1['Cleaned_Job_Description'].head())

print("Resumes After Preprocessing:")
print(df2['Cleaned_Text'].head())

```

```python
from sklearn.model_selection import train_test_split

resumes_train, resumes_test = train_test_split(df2['Cleaned_Text'].dropna().tolist(), test_size=0.2, random_state=42)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

job_vectors = model.encode(df1['Cleaned_Job_Description'].dropna().tolist())
resume_vectors_test = model.encode(resumes_test)
similarity_matrix_test = cosine_similarity(resume_vectors_test, job_vectors)

```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

def evaluate_model(similarity_matrix_test, threshold=0.7):
    y_true = []
    y_pred = []

    for similarities in similarity_matrix_test:
        if any(sim > threshold for sim in similarities):
            y_pred.append(1)
        else:
            y_pred.append(0)


        y_true.append(1)

    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)

    return precision, recall, f1, accuracy
```

```python
def visualize_similarity_distribution(similarity_scores, threshold=0.7):
    plt.figure(figsize=(10, 6))
    sns.histplot(similarity_scores, bins=30, kde=True, color='b')
    plt.axvline(x=threshold, color='r', linestyle='--', label=f'Threshold ({threshold})')
    plt.title("Distribution of Cosine Similarity Scores", fontsize=16)
    plt.xlabel("Cosine Similarity Score", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.legend()
    plt.show()
```
